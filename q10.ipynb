```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
publication = []
for pub in soup.find_all('td', class_ = "gsc_mvt_t"):
    publication.append(pub.text)
publication
```

```python
h5_index = []
for h in soup.find_all(class_ = "gs_ibl gsc_mp_anchor"):
    h5_index.append(h.text)
h5_index
```

```python
h5_median = []
for hm in soup.find_all(class_ = "gs_ibl gsc_mp_anchor"):
    h5_median.append(hm.text)
h5_median
```

```python
rank = []
for r in soup.find_all('td', class_="gsc_mvt_p"):
    rank.append(r.text)
rank
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Rank': rank, 'PUBLICATION': publication, 'H5_INDEX':h5_index, 'H5_MEDIAN': h5_median})
```

```python
print(len(rank), len(publication), len(h5_index), len(h5_median))
```
